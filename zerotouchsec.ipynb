```python
import cv2
from cvzone.HandTrackingModule import HandDetector
import os
import boto3

cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.8, maxHands=1)
```

```python
status, photo = cap.read()
print("Camera status:", status)

cv2.imshow("hi", photo)
cv2.waitKey()  
cv2.destroyAllWindows()
```

```python
hands, img = detector.findHands(photo)

if hands:
    hand = hands[0]
    fingers = detector.fingersUp(hand)
    print("Detected finger states:", fingers)

    if fingers == [1, 1, 1, 1, 1]:
        os.system("notepad")
    elif fingers == [0, 1, 1, 0, 0]:
        os.system("google")
    else:
        print("Gesture not mapped")
else:
    print("No hand detected")



```

```python
ec2 = boto3.client("ec2")

if fingers == [1, 1, 1, 1, 1]:  
    print("Launching EC2 instance...")
    response = ec2.run_instances(
        ImageId="ami-0c55b159cbfafe1f0", 
        InstanceType="t2.micro",
        MinCount=1,
        MaxCount=1,
    )
    instance_id = response['Instances'][0]['InstanceId']
    print("Launched instance:", instance_id)

elif fingers == [0, 1, 1, 0, 0]:  cap.release()
print("Camera released, notebook finished.")

    print("Setting up RDS...")
    rds = boto3.client("rds")
    response = rds.create_db_instance(
        DBName="mydb",
        DBInstanceIdentifier="gesture-db",
        AllocatedStorage=20,
        DBInstanceClass="db.t2.micro",
        Engine="mysql",
        MasterUsername="admin",
        MasterUserPassword="yourpassword123"
    )
    print("Created RDS:", response)
```
